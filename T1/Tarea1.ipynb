**1. (5 puntos) Normas.**  
Demuestre la siguiente identidad:

$$
\text{rms}_w^2(x) = \text{avg}_w^2(x) + \text{std}_w^2(x)
$$

donde $w_1, w_2, \dots, w_n$ son pesos positivos tales que la suma total de los pesos es:

$$
W = \sum_{i=1}^{n} w_i > 0
$$

y donde tenemos las definiciones:

$$
\text{avg}_w(x) = \frac{1}{W} \sum_{i=1}^{n} w_i x_i
$$

$$
\text{std}_w^2(x) = \frac{1}{W} \sum_{i=1}^{n} w_i (x_i - \text{avg}_w(x))^2
$$

$$
\text{rms}_w(x) = \sqrt{\frac{1}{W} \sum_{i=1}^{n} w_i x_i^2}
$$

**2. (15 puntos) Aplicación de algoritmo <code>k-means</code>.**  
En este problema trataremos el problema de *clustering*, es decir, determinar los \( k \) representantes de un conjunto de datos y las etiquetas de estos que minimizan la distancia de cada punto a su representante (centro de clúster).

**a)** Programe el algoritmo de <code>k-means</code>. Construya una rutina:

```
k_means_fit(X, Z0, NITERMAX)
```

donde <code>X</code> son los datos, <code>Z0</code> son los representantes iniciales, y <code>NITERMAX</code> es el número máximo de iteraciones.

<details>
<summary><strong>Ver algoritmo <code>k-means</code></strong></summary>

<div style="background-color:#f2f2f2; border:1px solid #888; border-radius:4px; padding:10px;">
<p><strong>Solución:</strong><br>
El algoritmo *k-means* busca particionar un conjunto \( \mathbf{X} = \{x_1, \dots, x_n\} \subset \mathbb{R}^d \) en \( k \) clústeres disjuntos \( C_1, \dots, C_k \), minimizando la suma de distancias cuadradas entre cada punto y su centroide correspondiente.</p>

<p><strong>Paso 1: Asignación de clústeres</strong></p>
$$
z_i^{(t)} = \arg\min_{j \in \{1, \dots, k\}} \|x_i - \mu_j^{(t)}\|^2
$$

<p><strong>Paso 2: Actualización de centroides</strong></p>
$$
\mu_j^{(t+1)} = \frac{1}{|C_j^{(t)}|} \sum_{x_i \in C_j^{(t)}} x_i
$$

<p><strong>Paso 3: Evaluación de la función objetivo</strong></p>
$$
J^{(t)} = \frac{1}{n} \sum_{i=1}^{n} \|x_i - \mu_{z_i^{(t)}}\|^2
$$
</div>
</details>

````python
def k_means_fit(X, Z0, NITERMAX):
    centroids = Z0.copy()
    objective_history = []

    for _ in range(NITERMAX):
        data = X[:, np.newaxis]
        # Cálculo de distancias
        distances = np.linalg.norm(data - centroids, axis=2)
        # Asignación de clústeres
        labels = np.argmin(distances, axis=1)

        for k in range(centroids.shape[0]):
            if np.any(labels == k):
                # Actualización de centroides
                centroids[k] = X[labels == k].mean(axis=0)

        mu_z = centroids[labels]
        # Evaluación de función objetivo
        current_objective = np.mean((X - mu_z) ** 2)

        objective_history.append(current_objective)

    return centroids, labels, objective_history
````
