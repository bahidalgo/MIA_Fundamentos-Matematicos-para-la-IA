$$
\text{rms}_w^2(x) = \text{avg}_w^2(x) + \text{std}_w^2(x)
$$

donde $w_1, w_2, \dots, w_n$ son pesos positivos tales que la suma total de los pesos es:

$$
W = \sum_{i=1}^{n} w_i > 0
$$

y donde tenemos las definiciones:

$$
\text{avg}_w(x) = \frac{1}{W} \sum_{i=1}^{n} w_i x_i
$$

$$
\text{std}_w^2(x) = \frac{1}{W} \sum_{i=1}^{n} w_i (x_i - \text{avg}_w(x))^2
$$

$$
\text{rms}_w(x) = \sqrt{\frac{1}{W} \sum_{i=1}^{n} w_i x_i^2}
$$

#**2. (15 puntos) Aplicación de algoritmo k-means</code>.**


En este problema trataremos el problema de *clustering*, es decir, determinar los \( k \) representantes de un conjunto de datos y las etiquetas de estos que minimizan la distancia de cada punto a su representante (centro de clúster).

**a)** Programe el algoritmo de <code>k-means</code>. Construya una rutina:

```
k_means_fit(X, Z0, NITERMAX)
```

donde <code>X</code> son los datos, <code>Z0</code> son los representantes iniciales, y <code>NITERMAX</code> es el número máximo de iteraciones.

<details>
<summary><strong>Ver algoritmo <code>k-means</code></strong></summary>

<div style="background-color:#f2f2f2; border:1px solid #888; border-radius:4px; padding:10px;">
<p><strong>Solución:</strong><br>
El algoritmo <em>k-means</em> busca particionar un conjunto \( \mathbf{X} = \{x_1, \dots, x_n\} \subset \mathbb{R}^d \) en \( k \) clústeres disjuntos \( C_1, \dots, C_k \), minimizando la suma de distancias cuadradas entre cada punto y su centroide correspondiente.</p>

<p><strong>Paso 1: Asignación de clústeres</strong></p>
$$
z_i^{(t)} = \arg\min_{j \in \{1, \dots, k\}} \|x_i - \mu_j^{(t)}\|^2
$$

<p><strong>Paso 2: Actualización de centroides</strong></p>
$$
\mu_j^{(t+1)} = \frac{1}{|C_j^{(t)}|} \sum_{x_i \in C_j^{(t)}} x_i
$$

<p><strong>Paso 3: Evaluación de la función objetivo</strong></p>
$$
J^{(t)} = \frac{1}{n} \sum_{i=1}^{n} \|x_i - \mu_{z_i^{(t)}}\|^2
$$
</div>
</details>


````python
def k_means_fit(X, Z0, NITERMAX):
    centroids = Z0.copy()
    objective_history = []

    for _ in range(NITERMAX):
        data = X[:, np.newaxis]
        # Cálculo de distancias
        distances = np.linalg.norm(data - centroids, axis=2)
        # Asignación de clústeres
        labels = np.argmin(distances, axis=1)

        for k in range(centroids.shape[0]):
            if np.any(labels == k):
                # Actualización de centroides
                centroids[k] = X[labels == k].mean(axis=0)

        mu_z = centroids[labels]
        # Evaluación de función objetivo
        current_objective = np.mean((X - mu_z) ** 2)

        objective_history.append(current_objective)

    return centroids, labels, objective_history
````

````python
def plot_clusters_grid(data, results, current_seed):
    sns.set_style("whitegrid")
    n_clusters = len(next(iter(results.values()))[0])
    colors = sns.color_palette("husl", n_clusters)
    n_results = len(results)
    n_cols = min(3, n_results)
    n_rows = int(np.ceil(n_results / n_cols))
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(5 * n_cols, 5 * n_rows))
    axes = np.array(axes).reshape(n_rows, n_cols)
    for idx, (n_iter, (centroids, labels)) in enumerate(results.items()):
        row = idx // n_cols
        col = idx % n_cols
        ax = axes[row, col]
        for k in np.unique(labels):
            cluster_points = data[labels == k]
            ax.scatter(
                cluster_points[:, 0],
                cluster_points[:, 1],
                color=colors[k],
                s=30,
                alpha=0.7,
                edgecolor='w',
                linewidth=0.5,
                label=f'Cluster {k}'
            )
        ax.scatter(
            centroids[:, 0],
            centroids[:, 1],
            c='black',
            marker='*',
            s=300,
            edgecolor='gold',
            linewidth=1.5,
            label='Centroids'
        )
        ax.set_title(f'Iteración {n_iter}', fontsize=13, fontweight='bold')
        ax.set_xlabel('Feature 0', fontsize=11)
        ax.set_ylabel('Feature 1', fontsize=11)
        ax.grid(True, linestyle='--', alpha=0.7)
        for spine in ax.spines.values():
            spine.set_edgecolor('#DDDDDD')
            spine.set_linewidth(1.2)
    # Eliminar subplots vacíos
    for i in range(n_results, n_rows * n_cols):
        fig.delaxes(axes.flatten()[i])
    # Leyenda personalizada
    legend_elements = [
        Line2D([0], [0], marker='o', color='w',
               label=f'Cluster {k}',
               markerfacecolor=colors[k],
               markersize=10)
        for k in range(n_clusters)
    ] + [
        Line2D([0], [0], marker='*', color='black',
               label='Centroids',
               markerfacecolor='gold',
               markersize=14,
               markeredgewidth=1.5)
    ]
    # Leyenda fuera del área del gráfico
    fig.legend(
        handles=legend_elements,
        loc='upper center',
        bbox_to_anchor=(0.5, 0.9),
        ncol=min(n_clusters + 1, 6),
        frameon=True,
        fontsize=10,
        borderpad=0.5,
        labelspacing=0.8,
        columnspacing=1.2
    )
    plt.tight_layout(rect=[0, 0, 1, 0.84])
    plt.show()
````

**d)** Discuta porque el algoritmo de `k−means` converge, es decir, ¿es
seguro que de una iteración a la siguiente la función objetivo decrece?

**Explicación:**  
Dado el tercer paso mencionado en la sección 2.a, el algoritmo *k-means*
garantiza la convergencia debido a que la función objetivo es
descendente.

**Recordando la evaluación de la función objetivo, se tiene:**

\$\$ J^{(t)} = \frac{1}{n} \sum\_{i=1}^{n} \\\|x_i -
\mu\_{z_i^{(t)}}\\\|^2 \$\$

Donde \$z_i\$ es:

\$\$ z_i^{(t)} = \arg\min\_{j \in \\{1, \dots, k\\}} \\\|x_i -
\mu_j^{(t)}\\\|^2 \$\$

Lo anterior implica que en cada punto \$x_i\$ se compara todas las
posibles distancias a los k posibles centroides. Luego, se escoge el
mínimo local del término: \$\$ \\\|x_i - \mu\_{z_i^{(t)}}\\\|^2 \$\$
respecto a \$z_i\$. Por ello, para la iteración siguiente repetiré el
proceso, buscando una asignación para \$z_i\$ mejor o igual, por ende:
\$\$ \\\|x_i - \mu\_{z_i^{(t)}}\\\|^2 \geq \\\|x_i - \mu\_{z_i^{(t +
1)}}\\\|^2 \$\$ Así, para cada iteración, se debe mantener la
desigualdad: \$\$ J^{t} \geq J^{t + 1} \\ \\ \\ \\ \forall t \in
\mathbb{N} \$\$ Por ello, aseguramos que la función objetivo decrece.
Sin embargo, esto no es suficiente para determinar que en efecto
converge, por ello, debemos encontrar una cota inferior, que permita que
el límite cuando tienda a dicha cota exista, lo cual sin embargo, es
evidente pues dado que la función objetivo es una suma de normas
cuadradas se tiene: \$\$ J^{t} \geq 0 \$\$ para todo \$t\$. Lo que
implica, junto a lo anterior, que la sucesión es monótona decreciente y
acotada inferiormente. Así se cumple: \$\$ \lim\_{t \to \infty} J^{(t)}
\in \mathbb{R} \$\$ Lo que es equivalente a indicar que la función
objetivo converge. Ahora bien, falta aclarar un último punto, dado que
la pregunta hace alusión a la convergencia del algoritmo y lo demostrado
anteriormente corresponde a la convergencia de la sucesión que no es lo
mismo. Para esto, es necesario añadir a lo anterior un hecho que permita
que el algoritmo se detenga eventualmente en un número finito de pasos.
Para lo anterior, basta con notar que cada punto \$x_i\$ se asigna a uno
de los \$k\$ clústeres posibles, lo que genera como máximo \$k^n\$
posibles asignaciones distintas. Dado que el algoritmo mejora la función
objetivo en cada paso, debe existir un número finito de iteraciones en
donde el algoritmo se detendrá.

Considere una matriz \\( A \in \mathbb{R}^{m \times n} \\) con \\( m \>
n \\) y un vector \\( b \in \mathbb{R}^m \\). Buscamos el vector
solución \\( x \in \mathbb{R}^n \\) del sistema lineal \\( Ax = b \\).
Describa cómo puede asegurar de forma práctica si la solución del
sistema existe o no.

Para el caso \\( n = 3 \\) y \\( m = 4 \\), considere el siguiente
sistema de ecuaciones:

\$\$ \begin{aligned} x + y + z &= 1, \\\\ 2x - 3y + 4z &= 1, \\\\ 3x +
5y - z &= 1, \\\\ 2y - 4z &= -2. \end{aligned} \$\$

<span class="arrow">▶️</span> Ver respuesta

Una forma práctica de determinar si un sistema \\( Ax = b \\) tiene
solución es verificar si el vector \\como( b \\) pertenece al *espacio
columna* de la matriz \\( A \\). Esto se puede hacer mediante la
reducción por filas de la matriz aumentada \\( \[A\\ \|\\ b\] \\),
comparando su rango con el de \\( A \\).

Según el [Teorema de Rouché–Capelli](#capelli1895) (también conocido
como Rouché-Frobenius), el sistema es resoluble con valores finitos de
las incógnitas si, y solo si, ambas matrices tienen el mismo rango
(Capelli, 1895):

-   Si \\( \text{rango}(A) = \text{rango}(\[A\\ \|\\ b\]) \\), el
    sistema es *consistente*.
-   Si \\( \text{rango}(A) \neq \text{rango}(\[A\\ \|\\ b\]) \\), el
    sistema es *inconsistente*.

**Matriz aumentada:**

\$\$ \left\[ \begin{array}{ccc\|c} 1 & 1 & 1 & 1 \\\\ 2 & -3 & 4 & 1
\\\\ 3 & 5 & -1 & 1 \\\\ 0 & 2 & -4 & -2 \end{array} \right\] \$\$

**Paso 1:** Pivote \\( A\_{11} = 1 \\). Multiplicadores:

\$\$ \ell\_{21} = \frac{2}{1}, \quad \ell\_{31} = \frac{3}{1}, \quad
\ell\_{41} = \frac{0}{1} \$\$

Operaciones:

\$\$ \begin{aligned} F_2 &\leftarrow F_2 - \ell\_{21} F_1 \\\\ F_3
&\leftarrow F_3 - \ell\_{31} F_1 \\\\ F_4 &\leftarrow F_4 - \ell\_{41}
F_1 \end{aligned} \$\$

**Resultado:**

\$\$ \left\[ \begin{array}{ccc\|c} 1 & 1 & 1 & 1 \\\\ 0 & -5 & 2 & -1
\\\\ 0 & 2 & -4 & -2 \\\\ 0 & 2 & -4 & -2 \end{array} \right\] \$\$

**Paso 2:** Pivote \\( A\_{22} = -5 \\). Multiplicadores:

\$\$ \ell\_{32} = \frac{2}{-5}, \quad \ell\_{42} = \frac{2}{-5} \$\$

Operaciones:

\$\$ \begin{aligned} F_3 &\leftarrow F_3 - \ell\_{32} F_2 \\\\ F_4
&\leftarrow F_4 - \ell\_{42} F_2 \end{aligned} \$\$

**Resultado:**

\$\$ \left\[ \begin{array}{ccc\|c} 1 & 1 & 1 & 1 \\\\ 0 & -5 & 2 & -1
\\\\ 0 & 0 & -\frac{16}{5} & -\frac{12}{5} \\\\ 0 & 0 & -\frac{16}{5} &
-\frac{12}{5} \end{array} \right\] \$\$

**Paso 3:** Pivote \\( A\_{33} = -\frac{16}{5} \\)

\$\$ \ell\_{43} = \frac{-\frac{16}{5}}{-\frac{16}{5}} = 1 \$\$

Operación:

\$\$ F_4 \leftarrow F_4 - \ell\_{43} F_3 \$\$

**Resultado final:**

\$\$ \left\[ \begin{array}{ccc\|c} 1 & 1 & 1 & 1 \\\\ 0 & -5 & 2 & -1
\\\\ 0 & 0 & -\frac{16}{5} & -\frac{12}{5} \\\\ 0 & 0 & 0 & 0
\end{array} \right\] \$\$

Observamos que el rango de \\( A \\) es 3 y el rango de la matriz
aumentada también es 3. Por lo tanto, como los rangos coinciden, el
sistema tiene solución.

Al resolver el sistema triangular por sustitución hacia atrás, se
obtiene como solución única: $ x = -\frac{1}{4}, \quad y =
\frac{1}{2}, \quad z = \frac{3}{4}. $



**4. (15 puntos) Clasificador binario: Perceptrón**

En este problema trataremos el problema de clasificación binaria, es
decir, determinar si un dato pertenece a una de dos clases.

**a)** Programe el algoritmo del Perceptrón. Puede ser útil construir en
este caso una rutina

    Perceptron_fit(X, y, nitmax, eta)

que tome los datos `X`, sus etiquetas `y`, y que retorne los pesos `w`
del clasificador. Otros hiperparámetros a considerar son `nitmax` y
`eta` que corresponden respectivamente al número máximo de iteraciones
del algoritmo y el *learning rate*.

El algoritmo del perceptrón, tal como se presenta en la clase 2 del
curso, incluye la siguiente ecuación para la actualización de pesos:
$ w^{(k+1)} = w^{(k)} - (y - \hat{y}) x_i, $ donde \$y \in \\{-1,
1\\}\$ es la etiqueta verdadera e \\(\hat{y} = \text{sgn}(x_i^\top w)\\)
es la predicción del modelo. Ahora, es posible notar que: \$\$ y \in
\\{-1, 1\\} \implies (y - \hat{y}) \in \\{-2, 0, 2\\} \$\$ En
consecuencia, si el ejemplo es clasificado correctamente (\\(y =
\hat{y}\\)), la corrección es nula; si es clasificado erróneamente (\\(y
\ne \hat{y}\\)), la corrección queda como: $ w^{(k+1)} = w^{(k)} - 2y
x_i, $ lo que nos permite reescribir la regla de actualización como:
$ w^{(k+1)} = w^{(k)}- \eta \cdot y \cdot x_i, $ absorbiendo el
factor 2 en el valor del parámetro \\(\eta\\) (*learning rate*).

````python
def Perceptron_fit(X, y, nitmax, eta):
    # Cada x ∈ ℝⁿ⁻¹ debe ser aumentado con una entrada unitaria.
    X_aug = np.hstack([X, np.ones((X.shape[0], 1))])
    # Inicializamos w = 0 ∈ ℝⁿ
    w = np.zeros(X_aug.shape[1])
    
    # Guardamos el historial de w
    historia = [w.copy()] 
    # Bucle principal: hasta nitmax o convergencia
    for _ in range(nitmax):
        actualizado = False
        for i in range(X_aug.shape[0]):
            # Calcular ŷ = sgn(xᵢᵀ w)
            y_pred = np.sign(np.dot(X_aug[i], w))
            if y_pred == 0:
                y_pred = -1  # por convención
            # Si está mal clasificado, actualizar pesos
            if y_pred != y[i]:
                # Aplicamos la regla de aprendizaje:
                # w(k+1) = w(k) − η⋅y⋅xi,
                w -= eta * y[i] * X_aug[i]
                actualizado = True
        historia.append(w.copy())
        if not actualizado:
            break  # convergió
    return w, historia
````

Este proceso se repite por un número máximo de iteraciones, o hasta que
los datos sean perfectamente clasificados (en caso de que sean
linealmente separables). Notar que por conveniencia, se aprovecha de
guardar el historial de pesos para ser usados más adelante.

**b)** Programe una rutina que dado un vector de pesos `w`, y un
conjunto de datos `X` prediga a qué conjunto de datos pertenece cada
dato.

Durante el entrenamiento, se evalúa cada ejemplo individual \\( x_i \in
\mathbb{R}^n \\) contra el vector de pesos \\( w \in \mathbb{R}^n \\)
utilizando el producto escalar: $ \hat{y}\_i =
\operatorname{sgn}(x_i^\top w), $ lo cual produce una única
predicción escalar para ese punto. En cambio, para esta sección,
aprovecharemos los datos de entrada agrupados en una matriz 

$ X =
\begin{bmatrix} x_1^\top \\\\ x_2^\top \\\\ \vdots \\\\ x_m^\top
\end{bmatrix} \in \mathbb{R}^{m \times n}, $ y se utiliza el producto
matriz-vector:

$ Xw = \begin{bmatrix} x_1^\top w \\\\ x_2^\top w \\\\
\vdots \\\\ x_m^\top w \end{bmatrix} \in \mathbb{R}^m. $ Esta
formulación está basada en la definición de producto matriz-vector y su
interpretación lineal indicada en la clase 2 del curso. Finalmente, se
obtiene el vector de predicciones: $ \hat{y} =
\operatorname{sgn}(Xw), $

````python
def Perceptron_predict(w, X):
    # Aumentar con entrada unitaria (sesgo)
    X_aug = np.hstack([X, np.ones((X.shape[0], 1))])
    
    # Aplicar la regla: ŷ = sgn(Xw)
    y_pred = np.sign(X_aug @ w)
    y_pred[y_pred == 0] = -1 
    return y_pred   
````

**c)** Programe una rutina que dado un vector de pesos `w`, un conjunto
de datos `X`y sus etiquetas `y`, entregue un score de que tan bien
clasificados están los datos según las predicciones hechas por el
clasificador.

Para este caso simplemente consideraremos el porcentaje de detecciones
correctas.

````python
def Perceptron_score(w, X, y):
    # Hacemos uso de la función creada en b)
    y_pred = Perceptron_predict(w, X)
    score = (y_pred == y).mean() * 100
    return score 
````

**5. (5 puntos) Descomposición en valores singulares.**

Muestre que para $ A \in \mathbb{R}^{m \times n} $ con valores
singulares $ \sigma_1, \ldots, \sigma_p $, donde $ p = \min\\{m,
n\\} $, su norma de Frobenius satisface la identidad:

$ \\\|A\\\|\_F = \sqrt{\sigma_1^2 + \cdots + \sigma_p^2}. $

Verifique que la identidad anterior se cumple para la siguiente matriz:

$ A = \begin{bmatrix} 4 & 1 & 0 \\\\ 1 & 4 & 1 \\\\ 0 & 1 & 4
\end{bmatrix} $

<span class="arrow">▶️</span> Ver demostración

Sea $ A \in \mathbb{R}^{m \times n} $ una matriz de rango $ p =
\min\\{m,n\\} $ con descomposición en valores singulares $ A = U
\Sigma V^T $ donde $ \Sigma \in \mathbb{R}^{m \times n} $
contiene en su diagonal los valores singulares $ \sigma_1, \ldots,
\sigma_p $

La norma de Frobenius de $ A $ está definida como: $
\\\|A\\\|\_F^2 = \sum\_{i=1}^m \sum\_{j=1}^n a\_{ij}^2. $ Notar que
lo anterior se obtiene elevando al cuadrado la expresión para la norma
definida en la clase 2. Luego, nuestro objetivo es demostrar que: $
\\\|A\\\|\_F^2 = \sigma_1^2 + \sigma_2^2 + \cdots + \sigma_p^2. $

Usando la descomposición $ A = \sum\_{k=1}^p \sigma_k u_k v_k^T $,
donde $ u_k \in \mathbb{R}^m $ y $ v_k \in \mathbb{R}^n $ son
vectores ortonormales, se tiene que cada entrada de $ A $ se puede
escribir como: $ a\_{ij} = \sum\_{k=1}^p \sigma_k (u_k)\_i (v_k)\_j.
$ 

Entonces: $ \\\|A\\\|\_F^2 = \sum\_{i=1}^m \sum\_{j=1}^n \left(
\sum\_{k=1}^p \sigma_k (u_k)\_i (v_k)\_j \right)^2. $

Expandimos el cuadrado: $ \\\|A\\\|\_F^2 = \sum\_{i=1}^m
\sum\_{j=1}^n \sum\_{k=1}^p \sum\_{\ell=1}^p \sigma_k \sigma\_\ell
(u_k)\_i (v_k)\_j (u\_\ell)\_i (v\_\ell)\_j. $ 

Reordenando las sumas:
$ \\\|A\\\|\_F^2 = \sum\_{k=1}^p \sum\_{\ell=1}^p \sigma_k
\sigma\_\ell \left( \sum\_{i=1}^m (u_k)\_i (u\_\ell)\_i \right) \left(
\sum\_{j=1}^n (v_k)\_j (v\_\ell)\_j \right). $

Dado que $ \\{u_k\\} $ y $ \\{v_k\\} $ son conjuntos
ortonormales, se cumple por lo visto en la clase 3 que: \$ u_k^T
u\_\ell = \begin{cases} 1 & \text{si } k = \ell \\\\ 0 & \text{si } k
\ne \ell \end{cases} \quad\text{y}\quad v_k^T v\_\ell = \begin{cases} 1
& \text{si } k = \ell \\\\ 0 & \text{si } k \ne \ell \end{cases} $

Por lo tanto, todos los términos con $ k \ne \ell $ se anulan, y
sólo sobreviven los términos con $ k = \ell $. $ \\\|A\\\|\_F^2 =
\sum\_{k=1}^p \sigma_k \sigma_k \left( \sum\_{i=1}^m (u_k)^2_i \right)
\left( \sum\_{j=1}^n (v_k)^2_j \right). $ $ \\\|A\\\|\_F^2 =
\sum\_{k=1}^p \sigma_k \sigma_k \left( 1 \right) \left( 1 \right). $

Lo que lleva a la siguiente expresión: $ \\\|A\\\|\_F^2 =
\sum\_{k=1}^p \sigma_k^2. $

Finalmente, tomando raíz cuadrada a ambos lados: $ \\\|A\\\|\_F =
\sqrt{\sigma_1^2 + \cdots + \sigma_p^2}, $ lo cual demuestra la identidad deseada.

<span class="arrow">▶️</span> Ver verificación de propiedad

Calculamos primero la norma de Frobenius directamente desde los
elementos de la matriz:

$ \\\|A\\\|\_F^2 = 4^2 + 1^2 + 0^2 + 1^2 + 4^2 + 1^2 + 0^2 + 1^2 +
4^2 = 52 \quad \Rightarrow \quad \\\|A\\\|\_F = \sqrt{52}.$

Luego, al calcular los valores singulares de $ A $ (usaremos
`np.linalg.svd()`), obtenemos:




````python
    import numpy as np

    A = np.array([
        [4, 1, 0],
        [1, 4, 1],
        [0, 1, 4]
    ])

    # Calcular norma de Frobenius directamente
    frobenius = np.linalg.norm(A, ord='fro')
    print(f"Norma de Frobenius: {frobenius:.4f}")

    # Calcular valores singulares
    sigma = np.linalg.svd(A, compute_uv=False)
    print("Valores singulares:", sigma)

    # Verificar identidad
    print("Suma de cuadrados de los σ_i:", np.sum(sigma**2))
          

````
